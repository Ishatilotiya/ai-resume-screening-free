```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
```

```python
def extract_info(resume):
    prompt = f"""
Extract skills, tools, and experience from this resume:

{resume}

Output:
Skills:
Tools:
Experience:
"""
    return generator(prompt, max_length=150)[0]['generated_text']
```

```python
def match_skills(resume_data, job):
    prompt = f"""
Compare resume with job description.

Resume:
{resume_data}

Job:
{job}

Give:
- Matching skills
- Missing skills
"""
    return generator(prompt, max_length=150)[0]['generated_text']
```

```python
def calculate_score(resume):
    score = 0
    keywords = ["python", "machine learning", "nlp", "deep learning", "sql"]

    resume = resume.lower()

    for word in keywords:
        if word in resume:
            score += 20

    return min(score, 100)
```

```python
def generate_explanation(score):
    if score >= 80:
        return "Strong match: Candidate meets most job requirements."
    elif score >= 50:
        return "Average match: Candidate meets some requirements."
    else:
        return "Weak match: Candidate lacks key required skills."
```

```python
def run_pipeline(resume, job):

    extracted = extract_info(resume)
    match = match_skills(extracted, job)
    score = calculate_score(resume)
    explanation = generate_explanation(score)

    return {
        "Extracted": extracted,
        "Match": match,
        "Score": score,
        "Explanation": explanation
    }
```

```python
job_description = "Looking for Python, Machine Learning, NLP, Deep Learning, SQL"

strong_resume = "Python Machine Learning NLP Deep Learning SQL 3 years experience"
average_resume = "Python SQL basic Machine Learning"
weak_resume = "Excel Word communication skills"
```

```python
print("🔵 STRONG")
print(run_pipeline(strong_resume, job_description))

print("\n🟡 AVERAGE")
print(run_pipeline(average_resume, job_description))

print("\n🔴 WEAK")
print(run_pipeline(weak_resume, job_description))
```
